# Dataset Exploration

Quick orientation for the panel — distribution stats over the 100 enriched transcripts.

```python
from ti.store import duckdb_store
import pandas as pd
con = duckdb_store.connect()
```

## Call-type distribution

```python
con.execute('SELECT call_type, COUNT(*) AS n, ROUND(AVG(duration_min),1) AS avg_min FROM transcripts GROUP BY call_type ORDER BY n DESC').df()
```

## Sentiment by call type (pre-tagged 1–5 score)

```python
con.execute('SELECT call_type, ROUND(AVG(sentiment_score),2) AS avg_score, ROUND(AVG(arc_slope),3) AS avg_arc_slope FROM transcripts GROUP BY call_type ORDER BY avg_score').df()
```

## Top 10 customers by mention count (post-normalization)

```python
con.execute("SELECT name, SUM(mention_count) AS mentions, COUNT(DISTINCT meeting_id) AS meetings FROM entities WHERE kind='customer' GROUP BY name ORDER BY mentions DESC LIMIT 10").df()
```

## Risk-signal counts by kind and severity

```python
con.execute('SELECT kind, severity, COUNT(*) AS n FROM risk_signals GROUP BY kind, severity ORDER BY kind, severity').df()
```
