Per-owner closure rate, computed by tracking each action item's distinctive tokens into later utterances. Conservative by design — see plan §1 Failure Mode #2 ('no citation, no claim'). Open-item counts are the load-bearing signal even when closure rate trends low.

```python
from ti.insights.action_lineage import all_owners_summary, lineage_for_owner
import pandas as pd
```

## Per-owner summary (top 10 by item count)

```python
rows = all_owners_summary(min_items=2)
rows_sorted = sorted(rows, key=lambda s: -s.total_items)[:10]
pd.DataFrame([
    {
        'owner': s.owner,
        'total_items': s.total_items,
        'closed': s.closed_items,
        'open': s.total_items - s.closed_items,
        'closure_rate': s.closure_rate,
    }
    for s in rows_sorted
])
```

## Drill-down: highest open-item owner

```python
headline = rows_sorted[0]
print(f'Owner: {headline.owner}')
print(f'  total : {headline.total_items}')
print(f'  closed: {headline.closed_items}')
print(f'  open  : {headline.total_items - headline.closed_items}')
print(f'  rate  : {headline.closure_rate:.0%}')
print()
print('Open items:')
for o in headline.open_items[:10]:
    print(f'  · {o.source_meeting_id[-8:]}  — {o.description[:90]}')
```
